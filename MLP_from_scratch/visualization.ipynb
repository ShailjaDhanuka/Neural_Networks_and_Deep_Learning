# Plot implementation results

```python
losses, accuracies, losses2, accuracies2 = pickle.load(open("self_metrics.pkl", "rb"))

print(losses)
```

```python
#figure parameters
size = 5
Y1 = losses
Y2 = losses2
X = [i+1 for i in range(len(Y1))]

#plot the scatter plot
fig, ax = plt.subplots(figsize=(4, 3), dpi=300)
ax.plot(X, Y1, label="train")

ax.plot(X, Y2, label="test")
ax.set_ylabel(r"Loss", fontsize=size)
ax.set_xlabel(r"Epochs", fontsize=size)
plt.xticks(fontsize=size)
plt.yticks(fontsize=size)
ax.grid()
ax.legend()

#fig.savefig("filename.png")
```
